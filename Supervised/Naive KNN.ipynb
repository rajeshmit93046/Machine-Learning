 # Naive Bayes Theorem

**-Naïve Bayes algorithm is a supervised learning algorithm, which is based on Bayes theorem and used for solving classification problems.**

**It is mainly used in text classification that includes a high-dimensional training dataset.**

**Naïve Bayes Classifier is one of the simple and most effective Classification algorithms which helps in building the fast machine learning models that can make quick predictions.**

**It is a probabilistic classifier, which means it predicts on the basis of the probability of an object.**

**Some popular examples of Naïve Bayes Algorithm are spam filtration, Sentimental analysis, and classifying articles.**

# Implementation of Naive Bayes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Social_Network_Ads.csv")
data.head()
```

```python
data.info()
```

```python
data.describe().T
```

```python
data.isna().sum()
```

```python
X=data.iloc[:,:-1].values  #Feature data
y=data.iloc[:,-1].values  #Target variable
```

```python
X.shape
```

```python
y.shape
```

**Splitting dataset**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print(X_train.shape)
X_train
```

```python
print(X_test.shape)
```

```python
y_train.shape
```

```python
y_test.shape
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

**Training the model on Training Dataset**

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(X_train,y_train)
```

**Predicting new result**

Passing the Age 30 year and salary is 87000

```python
classifier.predict(sc.transform([[30,87000]]))
```

Here '0' mean customer will not purchased the car.

**Predicting on test dataset result**

```python
y_pred=classifier.predict(X_test)
y_pred
```

 **We can see the difference between real values of test dataset and Prediction values**

```python
df=pd.DataFrame({
    "Y_test":y_test,
    "Y_pred":y_pred
})
```

```python
df
```

**Making Confusion matrix**

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
acc=accuracy_score(y_test,y_pred)
```

```python
print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
print("The accuracy score of KNN model:",acc)
```
